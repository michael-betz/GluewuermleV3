# Lets create some interesting patterns

```python
sPts = list()
sDlys = list()

def plotWaveform( samplePoints, sampleDelay=0, Nbit=13 ):
    sampleDelay += 1
    sPts.append( samplePoints )
    sDlys.append( sampleDelay )
    nSamples = len(samplePoints)
    tSeq = 1.0/30*nSamples*sampleDelay
    convertedSamples = (2**Nbit - 1) * samplePoints
    energySum = sum( convertedSamples ) * nSamples * sampleDelay / 10000
    convertedSamples = array( convertedSamples, dtype=int )
    close("all")
    figure(figsize=(8,3))
    tVect = linspace(0, tSeq, nSamples, False)
    plot( tVect, convertedSamples, "-o" )
    xlabel("time [s]")
    ylabel("Brightness")
    grid(True)
    text(0.9, 0.9, " nSamples = %6d \n tSeq [s] = %6.1f \n energySum = %6d"%(nSamples, tSeq, energySum), ha='right', va='top', transform=gca().transAxes)
    display(gcf())
    close("all")
```

```python
def createSourcecode( Nbit=13, fName="myWaves.h" ):
    allSamples = hstack(sPts)
    nSamples = len(allSamples)
    allConvertedSamples = (2**Nbit - 1) * allSamples
        
    with open(fName, "w") as f:
        f.write( "#ifndef MYWAVES_H_\n" )
        f.write( "#define MYWAVES_H_\n\n" )
        f.write( "#include <inttypes.h>\n" )
        f.write( "#include <avr/pgmspace.h>\n\n" )

        f.write( "static const uint16_t waveFormData[%d] PROGMEM = {\n"%(nSamples) )
        for i,s in enumerate( allConvertedSamples ):
            f.write( "0x%04x, "%s )
            if( ((i+1)%10) == 0 ):
                f.write( "\n" )
        f.write( " };\n\n" )

                
        f.write("typedef struct{\n")
        f.write("	uint16_t *samplePointer;			//Pointing to first sample\n")
        f.write("	uint16_t energyCost;				//How expensive is the sequence\n")
        f.write("	uint8_t length;					    //How long is the sequence\n")
        f.write("	uint8_t sampleDelay;			    //delay between samples (in 1/30 s)\n")
        f.write("}oneWaveform_t;\n")
        f.write("\n")
        f.write("#define N_WAVEFORMS %d\n"%len(sPts))
        f.write("static const oneWaveform_t waveformTable[ N_WAVEFORMS ] PROGMEM = {\n")
        startIndex = 0
        for i,samplePoints in enumerate(sPts):
            nSamples = len(samplePoints)
            sampleDelay = sDlys[i]
            tSeq = 1.0/30*nSamples*sampleDelay
            convertedSamples = (2**Nbit - 1) * samplePoints
            energySum = sum( convertedSamples ) * nSamples * sampleDelay / 10000
            f.write( "    { &waveFormData[%d], %d, %d, %d },\n"%( startIndex, energySum, nSamples, sampleDelay ) )
            startIndex += nSamples
        f.write( " };\n\n" )
        f.write( "#endif /* MYWAVES_H_ */\n" )
        
```

### Firefly sleepiness

```python
LBATT_THRESHOLD=			3300		            #[mV] go to deep sleep mode to save battery
HBATT_THRESHOLD=			4190		            #[mV] switch on LEDs to burn overvoltage
MAXIMUM_COFFEE_LEVEL=	     500
MAXIMUM_VOLTAGE= 	(HBATT_THRESHOLD - 50)			#Will output MAXIMUM_SLEEPINESS
MINIMUM_VOLTAGE= 	(3500)                          #Will output 0
vCap = linspace(4200, 3200, 200)
temp2  = (( vCap - MINIMUM_VOLTAGE )) * MAXIMUM_COFFEE_LEVEL
temp2 /= ( MAXIMUM_VOLTAGE - MINIMUM_VOLTAGE )
temp2 = array( temp2, dtype=int )
temp2 = temp2.clip(0, MAXIMUM_COFFEE_LEVEL)
close()
plot(vCap, temp2, "-o")
grid(True)
display(gcf())
```
